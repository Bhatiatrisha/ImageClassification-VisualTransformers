```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import timm

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Simple image transform
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # ViT expects 224x224
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# CIFAR-10 Dataset (only using 2 classes for simplicity)
classes_to_use = ['airplane', 'automobile']
class_indices = [0, 1]

full_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
subset_indices = [i for i, (_, label) in enumerate(full_dataset) if label in class_indices]
small_dataset = torch.utils.data.Subset(full_dataset, subset_indices)

train_loader = DataLoader(small_dataset, batch_size=32, shuffle=True)

# Load a small Vision Transformer model
model = timm.create_model('vit_tiny_patch16_224', pretrained=True, num_classes=10)
model.to(device)

# Replace head for 2-class classification
model.head = nn.Linear(model.head.in_features, 2)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# Training loop (just 2 epochs for demo)
for epoch in range(2):
    running_loss = 0.0
    for images, labels in train_loader:
        # Filter only selected classes
        indices = [i for i, label in enumerate(labels) if label in class_indices]
        images = images[indices]
        labels = torch.tensor([class_indices.index(labels[i]) for i in indices])

        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {running_loss:.4f}")

print("‚úÖ Training done.")
```

```python
# Load test dataset with same preprocessing
test_dataset = datasets.CIFAR10(root='./data', train=False, transform=transform)
test_subset_indices = [i for i, (_, label) in enumerate(test_dataset) if label in class_indices]
test_small_dataset = torch.utils.data.Subset(test_dataset, test_subset_indices)
test_loader = DataLoader(test_small_dataset, batch_size=32, shuffle=False)

# Evaluation
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        indices = [i for i, label in enumerate(labels) if label in class_indices]
        images = images[indices]
        labels = torch.tensor([class_indices.index(labels[i]) for i in indices])

        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"üîé Test Accuracy: {accuracy:.2f}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Helper: Show image
def imshow(img, title, correct=True):
    img = img / 2 + 0.5  # unnormalize
    npimg = img.cpu().numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.title(title, color='green' if correct else 'red')
    plt.axis('off')

# Get a batch of test images
dataiter = iter(test_loader)
images, labels = next(dataiter)

# Filter only the 2 classes and preprocess labels
indices = [i for i, label in enumerate(labels) if label in class_indices]
images = images[indices]
labels = torch.tensor([class_indices.index(labels[i]) for i in indices])

images, labels = images.to(device), labels.to(device)
outputs = model(images)
_, predicted = torch.max(outputs, 1)

# Plot first 6 predictions
plt.figure(figsize=(12, 6))
for i in range(6):
    plt.subplot(2, 3, i + 1)
    title = f"Pred: {classes_to_use[predicted[i]]}\nTrue: {classes_to_use[labels[i]]}"
    correct = predicted[i] == labels[i]
    imshow(images[i], title, correct)
plt.tight_layout()
plt.show()
```
